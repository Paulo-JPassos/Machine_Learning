```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from urllib.request import urlopen

# Carregar imagem
url = "https://raw.githubusercontent.com/opencv/opencv/master/samples/data/smarties.png"

resp = urlopen(url)
image = np.asarray(bytearray(resp.read()), dtype=np.uint8)

img = cv2.imdecode(image, cv2.IMREAD_COLOR)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Escala de cinza
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Segmentação simples
_, segmented = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(12,5))

plt.subplot(121)
plt.imshow(gray, cmap='gray')
plt.title("Imagem Original na escala cinza")

plt.subplot(122)
plt.imshow(segmented, cmap='gray')
plt.title("Segmentação")

plt.show()
```
